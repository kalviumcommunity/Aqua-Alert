# ⚙️ Jupyter Kernel Control & Execution Management

This notebook is an intentional demonstration of **Jupyter kernel control**. It showcases how the kernel manages execution state, how to safely interrupt long-running processes, and how restarting the kernel clears the session environment.

---
## 💾 Section 1: Normal Execution & State Retention

Jupyter Notebooks are stateful. When we execute a cell, the kernel retains the variables and functions defined in memory. Subsequent cells can then access this shared state.

```python
# Defining a variable to demonstrate state retention
x = 100
print(f"Checkpoint 1: Initial value of x set to {x}")
```

```python
# Updating the state from a new cell
x += 50
print(f"Checkpoint 2: Updated value of x is now {x} (Retained from previous cell!)")
```

---
## ⏹️ Section 2: Interrupting a Long-Running Process

Sometimes, a code cell may run unexpectedly long or go into an infinite loop. In such cases, we can **Interrupt the Kernel** to stop the current execution without losing the overall session state.

```python
import time

print("Starting a long-running process (20 seconds)...")
print("ACTION: During the video, please Interrupt the Kernel now!")

for i in range(20):
    time.sleep(1)
    print(f"Running... {i+1}/20")
```

---
## 🔄 Section 3: Restarting the Kernel

**Restarting the Kernel** wipes the entire session memory. All variables, imports, and state are cleared. This is useful for testing if your notebook is 'reproducible' from a clean slate.

```python
try:
    print(f"Current value of x: {x}")
except NameError:
    print("SUCCESS: The kernel was restarted and 'x' no longer exists in memory!")
```

---
## ✨ Conclusion

By mastering **Interrupt** and **Restart** commands, we gain complete control over our computational environment. This ensures that we can handle hanging processes and maintain a clean, reproducible execution flow.
