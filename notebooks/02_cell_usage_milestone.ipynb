This notebook is an intentional demonstration of **correct cell usage** in a Jupyter environment. It highlights the distinction between **Code cells** and **Markdown cells**, focusing on separating logical execution from narrative explanation.

- Use Markdown cells for clear sectioning and explanations.
- Use Code cells for simple Python operations.

## 🚀 Section 1: Variable Definition and Arithmetic

In this section, we define two integer variables and calculate their sum. This demonstrates how Markdown cells can serve as headers and explain the intent behind the upcoming code block.

```python
# Defining simple numerical variables
a = 25
b = 75

# Calculating the sum
total_sum = a + b

print(f"Value of a: {a}")
print(f"Value of b: {b}")
print(f"The total sum is: {total_sum}")
```

---
## 🐍 Section 2: Iterating Over Data Structures

Now, we will demonstrate a simple loop. We'll create a list of items and iterate through them to perform a basic transformation. This illustrates how Markdown cells provide high-level context before moving into logical operations.

```python
# Defining a list of numbers
numbers = [1, 2, 3, 4, 5]
squares = [n ** 2 for n in numbers]

print(f"Original sequence: {numbers}")
print(f"Squared sequence: {squares}")

# Iterating and printing results
for n, s in zip(numbers, squares):
    print(f"Item {n} squared: {s}")
```
